Завдання 2. Аналіз повноти відповідей респондентів



```python
# Отримуємо множину унікальних імен питань зі схеми
schema_questions = set(df1['qname'].unique())

# Отримуємо множину назв колонок основного датафрейму
data_columns = set(df.columns)

# Залишаємо лише ті, що є в обох списках
common_questions = list(schema_questions.intersection(data_columns))

print(f"Ми будемо перевіряти відповіді на {len(common_questions)} запитань.")

# Рахуємо тих, хто відповів на ВСІ вибрані запитання
final_count = df[common_questions].dropna().shape[0]

# Фшнальний вивод число
final_count
```

Завдання 3. Статистичний аналіз досвіду респондентів

```python
# Вибираємо колонку WorkExp
work_exp = df['WorkExp']

# 1. Середнє значення (mean)
mean_val = work_exp.mean()

# 2. Медіана (median)
median_val = work_exp.median()

# 3. Мода (mode) - повертає серію, тому беремо перший елемент [0]
mode_val = work_exp.mode()[0]

# Виведення результатів у зручному форматі
print("Міри центральної тенденції для досвіду роботи (WorkExp):")
print(f"Середнє значення: {mean_val:.2f} років")
print(f"Медіана:          {median_val} років")
print(f"Мода:             {mode_val} років")
```

Завдання 5. Визначення популярності Python



```python
# Шукаємо 'Python' у списку мов.
#na=False порожні клітинки (NaN),case=False  будь який  регістр (python чи Python)
python_filter = df['LanguageHaveWorkedWith'].str.contains('Python', na=False, case=False)

# Рахуємо кількість людей, які знають Python
python_users_count = python_filter.sum()

# Загальну кількість респондентів (завдання 1)
total_respondents = df.shape[0]

# Рахуємо відсоток
python_percentage = (python_users_count / total_respondents) * 100

# 5. Виводимо результат
print(f"Відсоток респондентів, які програмують на Python: {python_percentage:.1f}%")
```

Завдання 7. Географічний аналіз компенсації Python-розробників

```python
# Відбираємо лише тих, хто працював з Python
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
py_devs = df[df['worked_with_python'] == True].copy()

# Видаляємо рядки, де відсутнє значення зарплати (ConvertedCompYearly)
py_devs_clean = py_devs.dropna(subset=['ConvertedCompYearly'])

# Групуємо дані за країною (Country) та обчислюємо середню (mean) та медіанну (median) зарплату для кожної країни
py_progers_by_country = py_devs_clean.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])

# Виводимо результат у вигляді таблиці
py_progers_by_country
```

Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

```python
# Вибираємо три стобпці
top_paid_data = df[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]

# Видаляємо порожні значення зарплат
top_paid_data = top_paid_data.dropna(subset=['ConvertedCompYearly'])

# Сортуємо за спаданням (ascending=False), щоб "багатії" опинилися вгорі
top_paid_data = top_paid_data.sort_values(by='ConvertedCompYearly', ascending=False)

# Оновлюємо індекси (reset_index), старий індекс видаляємо та беремо 5 перших записів(топ-5 найбільш оплачуваних респондентів)
top_5_respondents = top_paid_data.reset_index(drop=True).head(5)

# Фінальна таблиця
top_5_respondents
```
